```python
deathAges = deaths_from_covid['Age group'].unique()
print(deathAges)
```

```python
deathAges = deaths_from_covid['Age group']
for age in deathAges:
    deathAges = deathAges.replace('Under 1 year', 1)
    deathAges = deathAges.replace('1-4 years', (1+4)/2)
    deathAges = deathAges.replace('5-14 years', (5+14)/2)
    deathAges = deathAges.replace('15-24 years', (15+24)/2)
    deathAges = deathAges.replace('25-34 years', (25+34)/2)
    deathAges = deathAges.replace('35-44 years', (35+44)/2)
    deathAges = deathAges.replace('45-54 years', (45+54)/2)
    deathAges = deathAges.replace('55-64 years', (55+64)/2)
    deathAges = deathAges.replace('65-74 years', (65+74)/2)
    deathAges = deathAges.replace('75-84 years', (75+84)/2)
    deathAges = deathAges.replace('85 years and over', (85+100)/2)
    deathAges = deathAges.replace('All ages', '0')
    deathAges = deathAges.replace('All Ages', '0')
print(deathAges)
```

```python
deaths_from_covid["Ages of Death"] = deathAges
```

```python
deaths_from_covid
```

```python
deaths_from_covid["Ages of Death"]=deaths_from_covid["Ages of Death"].dropna(0)
deaths_from_covid["Ages of Death"]
```

```python
deaths_from_covid
```

```python
deaths_from_covid["Ages of Death"]=deaths_from_covid["Ages of Death"].astype("int64")
deaths_from_covid
```

```python
bar_data=deaths_from_covid.groupby(["Ages of Death"]).count()["COVID-19 Deaths"]
bar_data
```

```python
bar_data.plot(kind="bar", figsize=(10,3))
plt.title("Ages of Death vs COVID-19 Deaths")
plt.ylabel("Number of COVID-19 Deaths")
plt.xlabel("Ages of Death")
plt.show()
plt.tight_layout()
```

```python
ages_of_death = [0, 1, 2, 9, 19, 29, 39, 49, 59, 69, 79, 92]
number_of_COVID_deaths = [210, 95, 100, 95, 51, 45, 65, 82, 94, 98, 99, 102 ]
x_axis = np.arange(len(number_of_COVID_deaths))
plt.bar(x_axis, number_of_COVID_deaths, color="b", align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ages_of_death)
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(number_of_COVID_deaths)+0.4)
plt.title("Ages vs Number of COVID deaths")
plt.xlabel("Ages")
plt.ylabel("Number of COVID deaths")
```

```python
groupby_age=deaths_from_covid.groupby(["Ages of Death", "COVID-19 Deaths"])
groupby_age_df=pd.DataFrame(groupby_age.size())
age=pd.DataFrame(groupby_age_df.groupby(["Ages of Death"]).count())
age.columns=["Total Deaths"]
age["Percentage of Ages"] = (100*(age["Total Deaths"]/ages["Total Deaths"].sum()))
age["Percentage of Ages"] = age["Percentage of Ages"]
age
```
