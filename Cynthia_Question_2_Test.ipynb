```python
unbin=[]
for ages in deaths_from_covid['Age group']:
    unbin.append(ages)
print(unbin)
```

```python
deathAges = deaths_from_covid['Age group'].unique()
print(deathAges)
```

```python
deathAges = deaths_from_covid['Age group']
for age in deathAges:
    deathAges = deathAges.replace('Under 1 year', 1)
    deathAges = deathAges.replace('1-4 years', (1+4)/2)
    deathAges = deathAges.replace('5-14 years', (5+14)/2)
    deathAges = deathAges.replace('15-24 years', (15+24)/2)
    deathAges = deathAges.replace('25-34 years', (25+34)/2)
    deathAges = deathAges.replace('35-44 years', (35+44)/2)
    deathAges = deathAges.replace('45-54 years', (45+54)/2)
    deathAges = deathAges.replace('55-64 years', (55+64)/2)
    deathAges = deathAges.replace('65-74 years', (65+74)/2)
    deathAges = deathAges.replace('75-84 years', (75+84)/2)
    deathAges = deathAges.replace('85 years and over', (85+100)/2)
    deathAges = deathAges.replace('All ages', 'NaN')
    deathAges = deathAges.replace('All Ages', 'NaN')
print(deathAges)
```

```python
deaths_from_covid["Ages of Death"] = deathAges
```

```python
deaths_from_covid.head()
```

```python
deaths_from_covid = deaths_from_covid.dropna()
```

```python
deaths_from_covid
```

```python
cv_deaths=deaths_from_covid["COVID-19 Deaths"].unique()
print(cv_deaths)
```

```python
covid_deaths = deaths_from_covid['COVID-19 Deaths']
for death in covid_deaths:
    covid_deaths = covid_deaths.replace('NaN', 0)
print(covid_deaths)
```

```python
deaths_from_covid["Deaths from COVID-19"] = covid_deaths
deaths_from_covid
```

```python
deaths_from_covid = deaths_from_covid.dropna()
deaths_from_covid
```

```python
covid_age=deaths_from_covid[["Ages of Death", "Deaths from COVID-19"]]
covid_age_df=pd.DataFrame(covid_age)
covid_age_df
covid_age_df=covid_age_df.dropna(how="any")
covid_ages_group=covid_age_df.groupby(['Ages of Death']).count()['Deaths from COVID-19']
covid_ages_group
```

```python
covid_age=deaths_from_covid[["Ages of Death", "COVID-19 Deaths"]]
covid_age_df=pd.DataFrame(covid_age)
covid_age_df
covid_age_df=covid_age_df.dropna(how="any")
covid_ages_group=covid_age_df.groupby(['Ages of Death']).count()['COVID-19 Deaths']
covid_ages_group
```

```python
covid_ages_group.plot(kind="bar", figsize=(10,3))
plt.title("Ages vs COVID-19 Deaths")
plt.ylabel("Ages of Death")
plt.xlabel("COVID-19 Deaths")
plt.show()
plt.tight_layout()
```
